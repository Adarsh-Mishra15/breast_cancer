```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data, columns =dataset.feature_names)
```

```python
df
```

```python
# Independence and Dependence features
x = df
y= dataset.target
```

```python
y
```

```python
# Train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.33, random_state = 42)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Post pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15, 12))
tree.plot_tree(treeclassifier, filled= True)
```

```python
# Post pruning with depth
treeclassifier = DecisionTreeClassifier(max_depth = 3)
treeclassifier.fit(x_train,y_train)
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_pred,y_test)
score
```

# Decision Tree Prepruning and Hyperparameter tuning for huge data

```python
parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier = DecisionTreeClassifier()
classifier = GridSearchCV(treeclassifier, param_grid = parameter, cv= 5, scoring = 'accuracy')
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.best_params_
```

```python
y_pred = classifier.predict(x_test)
```
